# Movies Correlation Project in Python

### Import the packages required

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

### Import Dataset

```python
df = pd.read_csv(r'C:\Users\fc\Downloads\movies.csv')
```

```python
df.head()
```

### Check Missing Values

```python

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(percent_missing*100)))
```

### Drop empty rows

```python
df = df.dropna(axis=0)

df.head()
```

```python

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(percent_missing*100)))
```

### Check Datatype

```python
df.dtypes
```

### Check Outliers

```python
df.boxplot(column=['gross'])
```

### Changing Datatype of required values

```python
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['gross'].astype('int64')

df.head()
```

### Check Duplicate Rows

```python
df.duplicated().value_counts()
```

### Visualizing Budget vs Gross Earning

```python
plt.figure(figsize=(18,10))
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
plt.figure(figsize=(18, 10))
plt.scatter(x=df['budget'], y=df['gross'])
plt.title("Budget vs Gross Earnings")
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earnings')
```

### Correlations between Score and Gross income

```python
plt.figure(figsize=(18, 10))
plt.scatter(x=df['score'], y=df['gross'])
plt.title("Score vs Gross Earnings")
plt.xlabel('Score of Film')
plt.ylabel('Gross Earnings')
```

### Correlation Matrix between all numeric columns

```python
df.corr()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

 ### Top 15 compaies by Gross revenue

```python
CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
